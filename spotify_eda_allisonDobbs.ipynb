-Boxplots (see below).  We are setting the danceability values to the x axis.
-The line down the middle of the blue rectangle is the median value, about .58.
-The left edge of the box is the quartile 1 value; the right edge of the box is the quartile 3 value.
-The width of the box is the interquartile range, or the quartile 3 value minus the quartile 1 value.
-The vertical lines out to the sides of the box are called the "whiskers" in a "box and whisker plot". They are the minimum and maximum values as determined by the interquartile range. The minimum value (on the left) is calculated by taking the quartile 1 value minus 1.5 times the interquartile range. The maximum value (on the right) is calculated by taking 1.5 times the interquartile range plus the quartile 3 value.
-Anything that falls outside of these min and max values are considered outliers. In this example, you can see that there are a few outliers near the minimum. 

```python
sns.boxplot(x=df['danceability'])
```

Finding out these values can help you see if your data has a normal distribution. If it doesn't, then you'll have to apply some kind of transformation so that the data falls along a normal distribution before you perform any statistical tests or models to the data.

Multivariate analysis looks at how two variables are related to each other. 

We can create a heat map but we have to extract only the quantitative variables, since this only works on numeric data. So, we get the column names.

```python
df.columns
df_corr = df[['popularity','duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

For this table, you can see the correlation between pairs of variables. You can see that the correlation between popularity and duration_ms is -.007101 which means that these two variables are not correlated at all.  

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr, annot=True, cmap='mako')  #you can annotate it with the correlation coefficients, and change the color. 'mako' makes it different shades of blue.
```

In this heatmap, a darker color means that two variables are more strongly negatively correlated, and a lighter color means that they are more positively correlated. Since acoustiness and energy are somewhat strongly negatively correlated at .73 (I think this means that when one goes up, the other goes down), we can do a pair plot that shows the relationship of two variables on a scatter plot, along with a histogram.

```python
sns.pairplot(df[['acousticness', 'energy']], height=5, aspect = 1.5)
```

The scatterplots are messy because there are too many points. So, to make it easier to read, you can take a sample with .sample(). You're making a pair plot like we did above, so you just copy and paste that code here and replace [['acousticness', 'energy']] with df_sample.

```python
df_sample = df[['acousticness', 'energy']].sample(n=5000) 
sns.pairplot(df_sample, height=5, aspect = 1.5)
```

In this pair plot, you can see that the negative correlation is easier to see and is not as jumbled as the one above. You can also remove the random_state=1 parameter and run it a few more times to see if you get the same plot.

You can also change the type of chart that shows the distribution of the variable, with the diag_kind='kde' parameter.

```python
df_sample = df[['acousticness', 'energy']].sample(n=5000) 
sns.pairplot(df_sample, height=5, aspect = 1.5, diag_kind='kde')
#This shows it as a density plot instead of a histogram.
```

Now we will look at how to add more variables to a pair plot. We change the color of a specific point.

```python
df['track_genre'].value_counts()

#We pick 5 of the 114 genres generated by the value_counts(); selecting from the df the items where the track_genre 'is in' this list of items.

df_genres = df[df['track_genre'].isin(['pop', 'folk', 'emo', 'punk-rock', 'electronic'])]

df_sample = df_genres.sample(n=5000, random_state=1)

sns.pairplot(df_sample[['acousticness', 'energy', 'track_genre']], height=5, aspect = 1.5, diag_kind='kde', hue='track_genre')
```

You want to look for clusters when you encode variables on a density plot. There doesn't appear to be many variable clusters but you can see that the punk-rock songs (in orange) are higher in energy and folk songs are higher in acoustiness, which corresponds with how those genres generally are.
